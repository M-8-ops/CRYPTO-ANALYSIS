```python
import pandas as pd
df = pd.read_csv(r"C:\Users\ADMIN\Documents\bull_bear_markets_btc.csv")
df.info()
```

```python
df.head(50)
```

```python
Converting date string format to datetime
df['datum'] = pd.to_datetime(df['datum'], format='%Y-%m-%d')
df['datum']
```

```python
#Cheking the day for the last entry
df.loc[30, 'datum'].day_name()
```

```python
#Finding the earliest entry
df['datum'].min()
```

```python
#Finding the latest entry
df['datum'].max()
```

```python
df
```

```python
#Performing a slice to isolate bear market
bear_m = df['2013-01':'2020-01']
```

```python
#Performing a slice to isolate bull market
bull_m = df['2020-01':'2021-07']
```

```python
%matplotlib inline
#Visualizing bear market
bear_m.plot()
```

```python
#visualizing the bull market
bull_m.plot()
```

```python
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

bear_m.plot()
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Bear Market')


plt.tight_layout()
plt.show()
```

```python
from datetime import datetime, timedelta
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates

plt.style.use('seaborn')

bull_m.plot()
plt.gcf().autofmt_xdate()

plt.title('Bitcoin Bull Market')


plt.tight_layout()
plt.show()
```
